```python
import telebot
import webbrowser
import os
from telebot import types
from dotenv import load_dotenv
load_dotenv()
bot = telebot.TeleBot(os.getenv('token'))




@bot.message_handler(commands=["start"])
def start(massage):
    button =  types.InlineKeyboardMarkup()
    button.add(types.InlineKeyboardButton("перейти в каталог.", url = 'https://madcats.ru/mad-cats/plyus-minus-kotik/'))
    bot.send_message(massage.chat.id, f"Здравствуйте, {massage.from_user.first_name}", reply_markup=button)


@bot.message_handler(commands=["help"])
def start(massage):
    bot.send_message(massage.chat.id, "Ну тут ни кто не поможет")

@bot.message_handler(content_types=["photo"])
def get_photo(massage):
    bot.reply_to( massage , "Классная фотка")

@bot.message_handler()
def info(massage):
    if massage.text.lower() == 'привет':
        bot.send_message(massage.chat.id, f"Здравствуйте, {massage.from_user.first_name}")
    elif massage.text.lower() == 'прибухнём':
        bot.send_message(massage.chat.id, f"а давай!")    

bot.polling(non_stop=True)
```
